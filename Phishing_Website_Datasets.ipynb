# **Importing logistic Regression on datasets**

```
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```
y=d['Result']
x=d.drop('Result',axis=1)
```

```
reg=LogisticRegression(random_state =10,C=1.0)
```

# **Creating Train_test_split on Datasets**

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=44)
mod=reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix, classification_report
```

# **accuracy score of Datasets for c=1**

```
m=reg.score(x_test,y_test)
m
```

# **Classification report**

```
r=classification_report(y_test,y_pred)
r
```

# **Confusion matrix of datasets**

```
s=confusion_matrix(y_test,y_pred)
s
```

```

a=np.arange(1,100,1);
y1=[]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=44)
for i in a:
  reg=LogisticRegression(random_state =10,C=i)
  mod=reg.fit(x_train,y_train)
  y_pred=reg.predict(x_test)
  m=reg.score(x_test,y_test)
  y1.append(m);
 
  
  

  
  
  
```
